## 1. Dataset
During the dataset processing for displaying and reviewing the data, if there are any missing or empty values, the following steps can be taken:
1. **Data Validation**: Ensure that the dataset is complete and accurate before analysis begins. Check for any missing or null values in the dataset.

2. **Handling Missing Data**:
    - **Remove Missing Values**: If the number of missing values is small, consider removing the records with missing data.

3. **Report Missing Data**: Document any missing or incomplete data to maintain transparency in the analysis process.

```python
# Reading data from the CSV file
file_path = '../covid_19_indonesia_time_series_all.csv'  # Replace with the location of your CSV file
df = pd.read_csv(file_path)

print("\nDisplaying the dataset:")
df
```

```python
# 1. Data Validation
def validate_data(df):
    missing_data = df.isnull().sum()
    print("Number of missing values in the dataset:\n", missing_data)
    return missing_data

# 2. Handling Missing Data
def handle_missing_data(df):
    # Remove rows with missing values if their number is small
    df_cleaned = df.dropna()
    return df_cleaned

# 3. Reporting Missing Data
def report_missing_data(missing_data):
    for column, count in missing_data.items():
        if count > 0:
            print(f"The column '{column}' has {count} missing values.")

# Calling the above functions
missing_data = validate_data(df)
df_cleaned = handle_missing_data(df)
report_missing_data(missing_data)

print("\nDataFrame after handling missing data:")
df_cleaned

```

## 2. Performing Data Comparison
- **Comparison by Year based on Total Deaths and Total Recovered**: Group the data by year and compare total deaths with total recovered.

- **Comparison by Year and Island based on Total Deaths and Total Recovered**: Group the data by year and island, then compare total deaths with total recovered.

- **Comparison by Year and Province based on Total Deaths and Total Recovered**: Group the data by year and province, then compare total deaths with total recovered.

- **Comparison by Year based on Total Cases**: Group the data by year and compare total cases.

```python
# Displaying column names in the DataFrame
print("Columns in the DataFrame:\n", df_cleaned.columns)

# If the 'Date' column exists, change its format
if 'Date' in df.columns:
    try:
        # Convert the 'Date' column to datetime format
        df['Date'] = pd.to_datetime(df['Date'])

        # Add a 'Year' column based on the 'Date' column
        df['Year'] = df['Date'].dt.year

        # Display only the 'Year' column and drop the 'Date' column if it's not needed
        df.drop(columns=['Date'], inplace=True)

        # Display the DataFrame with the new 'Year' column
        print(df.head())
    except Exception as e:
        print(f"An error occurred while processing the 'Date' column: {e}")
else:
    print("The 'Date' column was not found in the DataFrame.")
```

```python
# Performing Data Comparison
## Comparison by Year based on Total Deaths and Total Recovered
comparison_year = df_cleaned.groupby('Year')[['Total Deaths', 'Total Recovered']].sum().reset_index()

## Comparison by Year and Island based on Total Deaths and Total Recovered
comparison_year_island = df_cleaned.groupby(['Year', 'Island'])[['Total Deaths', 'Total Recovered']].sum().reset_index()

## Comparison by Year and Province based on Total Deaths and Total Recovered
comparison_year_province = df_cleaned.groupby(['Year', 'Province'])[['Total Deaths', 'Total Recovered']].sum().reset_index()

## Comparison by Year based on Total Cases
comparison_year_cases = df_cleaned.groupby('Year')['Total Cases'].sum().reset_index()

# Reporting the Results
print("\nComparison by Year based on Total Deaths and Total Recovered:\n", comparison_year)
print("\nComparison by Year and Island based on Total Deaths and Total Recovered:\n", comparison_year_island)
print("\nComparison by Year and Province based on Total Deaths and Total Recovered:\n", comparison_year_province)
print("\nComparison by Year based on Total Cases:\n", comparison_year_cases)

# Menyimpan DataFrame ke file teks dengan pemisah tab
comparison_year.to_csv('../Hasil/comparison_year.txt', sep='\t', index=False)
comparison_year_island.to_csv('../Hasil/comparison_year_island.txt', sep='\t', index=False)
comparison_year_province.to_csv('../Hasil/comparison_year_province.txt', sep='\t', index=False)
comparison_year_cases.to_csv('../Hasil/comparison_year_cases.txt', sep='\t', index=False)
```

```python
# Plotting Comparison by Year based on Total Deaths and Total Recovered
plt.figure(figsize=(12, 8))
sns.lineplot(data=comparison_year, x='Year', y='Total Deaths', label='Total Deaths')
sns.lineplot(data=comparison_year, x='Year', y='Total Recovered', label='Total Recovered')
plt.title('Comparison by Year based on Total Deaths and Total Recovered')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.show()

# Plotting Comparison by Year and Island based on Total Deaths
plt.figure(figsize=(12, 8))
sns.barplot(data=comparison_year_island, x='Year', y='Total Deaths', hue='Island')
plt.title('Comparison by Year and Island based on Total Deaths')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.show()

# Plotting Comparison by Year and Island based on Total Recovered
plt.figure(figsize=(12, 8))
sns.barplot(data=comparison_year_island, x='Year', y='Total Recovered', hue='Island')
plt.title('Comparison by Year and Island based on Total Recovered')
plt.xlabel('Year')
plt.ylabel('Total Recovered')
plt.show()

# Plotting Comparison by Year and Province based on Total Recovered
plt.figure(figsize=(12, 8))
sns.barplot(data=comparison_year_province, x='Year', y='Total Recovered', hue='Province')
plt.title('Comparison by Year and Province based on Total Recovered')
plt.xlabel('Year')
plt.ylabel('Total Recovered')
plt.show()

# Plotting Comparison by Year and Province based on Total Deaths
plt.figure(figsize=(12, 8))
sns.barplot(data=comparison_year_province, x='Year', y='Total Deaths', hue='Province')
plt.title('Comparison by Year and Province based on Total Deaths')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.show()

# Plotting Comparison by Year based on Total Cases
plt.figure(figsize=(12, 8))
sns.lineplot(data=comparison_year_cases, x='Year', y='Total Cases', label='Total Cases')
plt.title('Comparison by Year based on Total Cases')
plt.xlabel('Year')
plt.ylabel('Total Cases')
plt.legend()
```
